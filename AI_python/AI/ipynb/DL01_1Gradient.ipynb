```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b_target=1
w_target=2
x_data=[-10,-8,-5,-2,0,1,4,6,8,12]
y_data=np.zeros(10)
for i in range(len(x_data)):
  y_data[i]=w_target*x_data[i]+b_target
#y_data=w*x_data+b  ,
```

```python
y_data
```

```python
x_axis=np.arange(-5,5,0.1)#bias
y_axis=np.arange(-5,5,0.1)#weight
Z=np.zeros((len(x_axis),len(y_axis)))
X,Y=np.meshgrid(x_axis,y_axis)
```

```python
for i in range(len(x_axis)):
	for j in range(len(y_axis)):
		b=x_axis[i]
		w=y_axis[j]
		Z[j][i]=0
		for n in range(len(x_data)):
			y_hat=w*x_data[n]+b
			Z[j][i]=Z[j][i]+(y_data[n]-y_hat)**2
		Z[j][i]=Z[j][i]/len(x_data)
```

```python
plt.contourf(X,Y,Z,50,alpha=0.5,cmap=plt.get_cmap('jet'))
plt.xlabel(r'$b$',fontsize=16)
plt.ylabel(r'$w$',fontsize=16)
plt.plot(b_target,w_target,'x',ms=12,markeredgewidth=3,color='orange') #y_data=w*x_data+b  , w=2, b=1
```

```python
#ydata=b+w*xdata
w=-4	# initial w
b=-4  # initial b
lr=0.001 #lr=0.000001 #learning rate
iteration=200

b_history=[b]
w_history=[w]

for i in range(iteration):
	b_grad=0.0
	w_grad=0.0
	for n in range(len(x_data)):
		b_grad=b_grad -2.0*(y_data[n]-b-w*x_data[n])*1.0
		w_grad=w_grad -2.0*(y_data[n]-b-w*x_data[n])*x_data[n]

	b=b-lr*b_grad
	w=w-lr*w_grad

	b_history.append(b)
	w_history.append(w)
plt.contourf(X,Y,Z,50,alpha=0.5,cmap=plt.get_cmap('jet'))
plt.plot(b_target,w_target,'x',ms=12,markeredgewidth=3,color='orange')
plt.plot(b_history,w_history,'o-',ms=3,lw=1.5,color='black')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.xlabel(r'$b$',fontsize=16)
plt.ylabel(r'$w$',fontsize=16)
plt.show()
```
